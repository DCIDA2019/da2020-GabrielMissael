```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
```

```python
data = np.load("../datos/fit_exercise.npy")
```

```python
plt.errorbar(data[0], data[1], yerr= data[2])
plt.grid()
```

```python
def chi_square(P):
    total = 0
    for i in range(np.size(data[0])):
        total += (data[1][i]-(P[0]*data[0][i]+P[1]))**2/data[2][i]**2
    return total
```

```python
print(chi_square([-1, 6]))
```

```python
#DATOS EQUIDISTRIBUIDOS EN MALLA PARA M Y B
m_min = -2
m_max = 0
b_min = 2
b_max = 8
N = 100

m_inter = np.linspace(m_min, m_max, N)
b_inter = np.linspace(b_min, b_max, N)
chi = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        chi[i][j] = chi_square([m_inter[i], b_inter[j]])
        
print("Chi minima = ",np.amin(chi))
pos = np.where(chi == np.amin(chi))
m, b = m_inter[pos[0][0]], b_inter[pos[1][0]]
print("m = ", m, "b =", b)
```

```python
X, Y = np.meshgrid(m_inter, b_inter)
Z = chi
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100)
ax.set_xlabel('m (pendiente)')
ax.set_ylabel('b (ordenada al origen)')
ax.set_zlabel('chi-square')
ax.view_init(20, -40)
```

```python
plt.imshow(chi, cmap='hot')
plt.colorbar()
plt.xlabel("b (Ordenada al origen)")
plt.ylabel("m (Pendiente)")
#xticks(ticks, , **kwargs)
plt.show()
```

```python
def binaria_m(b):
    m_min = -2
    m_max = 0
    b_min = 2
    b_max = 8
    m_mid = (m_max-m_min)/2
    pos_max = chi_square([m_max, b])
    pos_min = chi_square([m_min, b])
    pos_mid = chi_square([m_mid, b])
    
    count = 0
    while(min(abs(m_mid-m_max), abs(m_mid-m_min))>0.00000000000001 and count<10000):
        count+=1
        if(pos_mid-pos_max<pos_mid-pos_min):
            m_min = m_mid
            pos_min = pos_mid
            m_mid = (m_max-m_min)/2
            pos_mid = chi_square([m_mid, b])
        else:
            m_max = m_mid
            pos_max = pos_mid
            m_mid = (m_max-m_min)/2
            pos_mid = chi_square([m_mid, b])
    
    return m_mid
```

```python
binaria_m(5.333333333333333)
```
