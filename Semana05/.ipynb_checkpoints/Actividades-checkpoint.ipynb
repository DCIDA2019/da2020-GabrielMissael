```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
from random import random
from scipy.stats import norm
from scipy import special
from mpl_toolkits.mplot3d import axes3d
```

```python
#Creamos las funciones de CDF y CDF inversa, esta segunda para dado una probabilidad dada por un numero aleatorio
#obtener su correspondiente valor
def CDF_normalinv(y, mu, sigma):
    return special.erfinv(2*y-1)*sigma*2**0.5+mu

def CDF_normal(y, mu, sigma):
    return 0.5*(1+special.erf((y-mu)/sigma*(2**0.5)))
```

```python
#Declaramos el numero de datos, el promedio (mu) y la desviacion (sigma)

N = 200000
x = np.sort(np.random.rand(N))
```

```python
#En "x" generamos N datos con distribución uniforme
fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(5,5))

ax0.hist(x, 50,  cumulative = False)
ax1.hist(x, 50, cumulative = True)
ax0.grid()
ax1.grid()
```

```python
fig, (ax0, ax1, ax3) = plt.subplots(3,1,figsize = (5,7))

ax0.plot(CDF_normalinv(x, 70, 2), x)
ax0.grid()
ax0.set_title("CDF de los datos aleatorios")

ax1.hist(CDF_normalinv(x, 70, 2), 50)
ax1.grid()

np.random.seed(5)
ax3.hist(np.random.randn(N)*2+70, 50, color = "orange")
ax3.grid()

plt.title("Distribución normal")
plt.show()
```

```python
def PDF_chi(x, k):
    return (x**(k/2-1)*mt.e**(-x/2))/(2**(k/2)*special.gamma(k/2))

def CDF_chi(x, k):
    return special.gammainc(k/2, x/2)/special.gamma(k/2)
```

```python
x = np.linspace(0.0001, 15, 1000)

fig, (ax0, ax1) = plt.subplots(1, 2,figsize = (12,4))

ax0.plot(x, PDF_chi(x, 1), label ="k=1")
ax0.plot(x, PDF_chi(x, 2), label ="k=2")
ax0.plot(x, PDF_chi(x, 3), label ="k=3")
ax0.plot(x, PDF_chi(x, 4), label ="k=4")
ax0.plot(x, PDF_chi(x, 5), label ="k=5")
ax0.plot(x, PDF_chi(x, 6), label ="k=6")
ax0.plot(x, PDF_chi(x, 7), label ="k=7")
ax0.plot(x, PDF_chi(x, 8), label ="k=8")
ax0.set_ylim(-0.05, 0.6)
ax0.set_xlim(-0.5, 9)
ax0.legend()
ax0.set_title("PDF")
ax0.grid()

ax1.plot(x, CDF_chi(x, 1), label ="k=1")
ax1.plot(x, CDF_chi(x, 2), label ="k=2")
ax1.plot(x, CDF_chi(x, 3), label ="k=3")
ax1.plot(x, CDF_chi(x, 4), label ="k=4")
ax1.plot(x, CDF_chi(x, 5), label ="k=5")
ax1.plot(x, CDF_chi(x, 6), label ="k=6")
ax1.plot(x, CDF_chi(x, 7), label ="k=7")
#ax1.set_ylim(-0.05, 1)
#ax1.set_xlim(-0.05, 9)
ax1.legend()
ax1.set_title("CDF")
ax1.grid()

plt.suptitle("Distribución Chi-cuadrada")
plt.show()
```

```python
#Creamos puntos aleatorios uniformes dentro de circulo de radio r centrado en el origen

N = 2000
R = 10
np.random.seed(5)

X = np.zeros(N)
Y = np.zeros(N)

pos = 0
while pos < N:
    x = np.random.rand(1)*2*R-R
    y = np.random.rand(1)*2*R-R
    if (x**2 + y**2) <= R**2:
        X[pos] = x
        Y[pos] = y
        pos+=1

fig, ax = plt.subplots(figsize=(5, 5))
plt.grid()
plt.scatter(X, Y, marker = ".")
```

```python
#Creamos puntos aleatorios uniformes dentro de esfera de radio r centrado en el origen

N = 3000
R = 10
np.random.seed(5)

X = np.zeros(N)
Y = np.zeros(N)
Z = np.zeros(N)

pos = 0
while pos < N:
    x = np.random.rand(1)*2*R-R
    y = np.random.rand(1)*2*R-R
    z = np.random.rand(1)*2*R-R
    if x**2 + y**2 + z**2 <=R**2:
        X[pos] = x
        Y[pos] = y
        Z[pos] = z
        pos+=1

fig = plt.figure(figsize = (7,7))
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(X, Y, Z, "r", marker='.')
```
