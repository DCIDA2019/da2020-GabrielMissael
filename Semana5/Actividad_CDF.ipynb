```python
N = 2000
R = 10
np.random.seed(5)

X = np.zeros(N)
Y = np.zeros(N)

pos = 0
while pos < N:
    x = np.random.rand(1)*2*R-R
    y = np.random.rand(1)*2*R-R
    if (x**2 + y**2) <= R**2:
        X[pos] = x
        Y[pos] = y
        pos+=1

fig, ax = plt.subplots(figsize=(5, 5))
plt.scatter(X, Y, marker = ".")
```

```python
N = 3000
R = 10
np.random.seed(5)

X = np.zeros(N)
Y = np.zeros(N)
Z = np.zeros(N)

pos = 0
while pos < N:
    x = np.random.rand(1)*2*R-R
    y = np.random.rand(1)*2*R-R
    z = np.random.rand(1)*2*R-R
    if x**2 + y**2 + z**2 <=R**2:
        X[pos] = x
        Y[pos] = y
        Z[pos] = z
        pos+=1

fig = plt.figure(figsize = (7,7))
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(X, Y, Z, "r", marker='.')
```
