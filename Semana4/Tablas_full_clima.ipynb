```python
import pandas as pd
import numpy as np
```

```python
T = []
TM = []
Tm = []
L = []
for i in range(9):
    T.append(pd.read_csv("../datos/Clima_full/T2019_0"+str(i+1)+".csv", encoding = 'Latin-1'))
    L.append(pd.read_csv("../datos/Clima_full/L2019_0"+str(i+1)+".csv", encoding = 'Latin-1'))
    TM.append(pd.read_csv("../datos/Clima_full/TM2019_0"+str(i+1)+".csv", encoding = 'Latin-1'))
    Tm.append(pd.read_csv("../datos/Clima_full/Tmin2019_0"+str(i+1)+".csv", encoding = 'Latin-1'))
    
T.append(pd.read_csv("../datos/Clima_full/T2019_10.csv", encoding = 'Latin-1'))
L.append(pd.read_csv("../datos/Clima_full/L2019_10.csv", encoding = 'Latin-1'))
TM.append(pd.read_csv("../datos/Clima_full/TM2019_10.csv", encoding = 'Latin-1'))
Tm.append(pd.read_csv("../datos/Clima_full/Tmin2019_10.csv", encoding = 'Latin-1'))
T.append(pd.read_csv("../datos/Clima_full/T2019_11.csv", encoding = 'Latin-1'))
L.append(pd.read_csv("../datos/Clima_full/L2019_11.csv", encoding = 'Latin-1'))
TM.append(pd.read_csv("../datos/Clima_full/TM2019_11.csv", encoding = 'Latin-1'))
Tm.append(pd.read_csv("../datos/Clima_full/Tmin2019_11.csv", encoding = 'Latin-1'))
```

```python
for i in range(11):
    T[i] = T[i].rename(columns={"Tmed":"Tmed_"+str(i+1)})
    TM[i] = TM[i].rename(columns={"Tmax":"Tmax_"+str(i+1)})
    Tm[i] = Tm[i].rename(columns={"Tmin":"Tmin_"+str(i+1)})
    print(Tm[i].columns)
```

```python
Tmed = pd.merge(T[0], pd.merge(T[1], pd.merge(T[2], pd.merge(T[3], pd.merge(T[4], pd.merge(T[5], 
              pd.merge(T[6], pd.merge(T[7], pd.merge(T[8], pd.merge(T[9], T[10]))))))))))
Tmax = pd.merge(TM[0], pd.merge(TM[1], pd.merge(TM[2], pd.merge(TM[3], pd.merge(TM[4], pd.merge(TM[5], 
              pd.merge(TM[6], pd.merge(TM[7], pd.merge(TM[8], pd.merge(TM[9], TM[10]))))))))))
Tmin = pd.merge(Tm[0], pd.merge(Tm[1], pd.merge(Tm[2], pd.merge(Tm[3], pd.merge(Tm[4], pd.merge(Tm[5], 
              pd.merge(Tm[6], pd.merge(Tm[7], pd.merge(Tm[8], pd.merge(Tm[9], Tm[10]))))))))))
```

```python
Temperatura = pd.merge(Tmed, pd.merge(Tmin, Tmax))
Temperatura
```
