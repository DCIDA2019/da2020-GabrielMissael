Queremos encontrar $\alpha$ y $\beta$ de la siguiente expresión:

$\mu = m_B* - (M_B - \alpha \times X_1 + \beta \times C) $

```python
datos = pd.read_csv("../datos/Supernova/data/jla_lcparams.txt", sep=" ")
datos.head()
```

```python
def mu(a, b, M):
    mu = np.zeros(740)
    for i in range(740):
        mu[i] = datos.iloc[i]["mb"] - (M - a*datos.iloc[i]["x1"] + b*datos.iloc[i]["color"])
    return mu

def grafica(x, y):
    fig, (ax0, ax1) = plt.subplots(2, 1, figsize = (7, 7))
    fig.subplots_adjust(hspace=0)
    
    ax0.scatter(x, y)
    ax0.grid()
    ax0.set_ylabel("$\mu$")
    ax0.set_xlabel("z")
    ax0.set_xlim(0.005, 2)
    ax0.set_xscale("log")
    
    plt.show()
```

```python
x = datos["zcmb"]

grafica(x, mu(0.1, 3, -19))
```
