```python
import pandas as pd
import numpy as np

#Cargamos los archivos .csv a un DataFrame
tempMax = pd.read_csv("temperatura_maxima.csv", encoding = 'Latin-1')
tempMid = pd.read_csv("temperatura_media.csv", encoding = 'Latin-1')
tempMin = pd.read_csv("temperatura_minima.csv", encoding = 'Latin-1')
lluvia18 = pd.read_csv("lluvia2018.csv", encoding = 'Latin-1')
lluvia19 = pd.read_csv("lluvia2019.csv", encoding = 'Latin-1')
lluvia20 = pd.read_csv("lluvia2020.csv", encoding = 'Latin-1')
```

```python
#Hacemos merge de las tres tablas. Usamos how = left para evitar que quite filas 
#donde falten entradas y mejor los llene con un NaN
temperatura = pd.merge(tempMax, pd.merge(tempMin, tempMid, how = 'left'), how = 'left')

#Creamos un indice multiple, de estado y localidad. Usamos sort.index() para ordenar 
#los estados por orden alfabetico
temperatura = temperatura.set_index(['Edo', 'Est', 'Clave']).sort_index(level = 0)

#Cambiamos los nombres de las columnas del indice multiple
temperatura = temperatura.rename_axis(['Estado', 'Localidad', 'Clave'])

#Previsualizamos la tabla
display(temperatura)
```

```python
temperatura.to_csv("TEMPERATURAS2019.csv")
```

```python
#Renombramos nombres de columnas para que coincidan y hacer el merge
#Pudimos haber indicado que son lo mismo sin cambiar el nombrem pero
#lo hacemos por si las tenemos que manipular de nuevo
lluvia19 = lluvia19.rename(columns={"EDO":"ESTADO"})
lluvia19 = lluvia19.rename(columns={"ESTACION":"NOMBRE"})
lluvia19 = lluvia19.rename(columns={"JUL":"2019"})

lluvia18 = lluvia18.rename(columns={"CLAVE_SIH":"CLAVE"})
lluvia18 = lluvia18.rename(columns={"ENE":"2018"})

lluvia20 = lluvia20.rename(columns={"CLAVE_SIH":"CLAVE"})
lluvia20 = lluvia20.rename(columns={"ENE":"2020"})
```

```python
#Repetimos lo mismo que con la tabla de temperaturas
lluvias = pd.merge(lluvia20, pd.merge(lluvia18, lluvia19, how = 'left'), how = 'left')
lluvias = lluvias.set_index(['ESTADO', 'NOMBRE', 'CLAVE']).sort_index()
```
